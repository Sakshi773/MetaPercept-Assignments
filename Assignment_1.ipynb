# *IMPORTING REQUIRED LIBRARIES*

```python
import chromadb
from sentence_transformers import SentenceTransformer
import pandas as pd
import string
```

```python
# taking sample documents

docs=[ " My name is Sakshi!",
       " I am currently pursuing masters in data science.",
       " During my academic years, I have acquired skills such as Python and SQL."
       ]


# preprocessing the documents
def cleaning_text(text):

  #converting to lowercase
  text=text.lower()

  #removing punctuation
  cleaned_text=""
  for i in text:
    if i not in string.punctuation:
      cleaned_text=cleaned_text+i

  return cleaned_text

# applying this preprocessing to all the documents
cleaned_docs=[]
for doc in docs:
  cleaned_docs.append(cleaning_text(doc))


print(cleaned_docs)



```

# *APPLYING EMBEDDING MODEL*

```python
model= SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(cleaned_docs)


```

# *STORING THE EMBEDDINGS IN CHROMADB*

```python
client= chromadb.Client()
collection=client.create_collection(name="my_collection")
collection=client.get_collection(name="my_collection")
```

```python
import numpy as np

id=[]
for i in range(len(cleaned_docs)):
  id.append(str(i))

#converting embedding to list
embeddings_list=embeddings.tolist()

# adding to the collection

collection.add(
    documents=cleaned_docs,
    embeddings=embeddings_list,
    ids=id
)

# Taking user query
query=input(" ASK ANYTHING:")
query=cleaning_text(query)
query_embedd=model.encode([query])[0]

# getting relevant documents
result=collection.query(
    query_embeddings=[query_embedd],
    n_results=1
)

# printing results
print("RESULTS:", result['documents'][0][0])




```
